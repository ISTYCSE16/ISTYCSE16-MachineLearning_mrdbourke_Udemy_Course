## It is recommended to use Object-Oriented Plotting instead of Pyplot, because Pyplot is generally less flexible

```python
# 1st Method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd Method
fig = plt.figure() # Creates a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # add some data
plt.show()
```

```python
# 3rd Method (recommended)
fig, ax = plt.subplots()
ax.plot(x, y); # add some data
```

# Anatomy Of A Matplotlib Figure

### Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in Colab
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot the data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title="Simple Plot", xlabel='x-axis', ylabel='y-axis')

# 5. Save & Show (you save the whole figure)
fig.savefig("..") 
```

**Bar Plotting**

```python
# Make a plot from Dictionary
nut_butter_prices = {"Almond Butter": 10,
                    "Peanut Butter": 8,
                    "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # Can't put (x, y)
# Because on the first argument you cannot put any positional argument. 
# On the second one you can put 'height' but can't put "x = nut_butter_prices()"
# so alternatively you can write "ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())"
ax.set(title="Dan's Nut Butter Store",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

**Histogram**

```python
# Make Some Data for histograms and plot it
x = np.random.randn(100)
fig, ax = plt.subplots()
ax.hist(x);
```

# Subplots


### Two Options

Subplot Option 1

```python
# Subplot Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

Subplot Option 2

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2, 
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

Which one should you use? (pyplot vs matplotlib OO Method?)
* When plotting something quickly, okay to use pyplot
* When plotting something more advanced, always use Object Oriented Method


```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot Method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# Object Oriented Method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

#ax.set_xlim([45, 100])
```

```python
# Set The Style
plt.style.use('seaborn-whitegrid')

# Object Oriented Method From Scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot The Data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap="plasma"); # This changes the color schemes

# Customize
ax.set(title="Heart Disease and Cholesterol Level",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add A Legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--')
```

```python
# Customizing the x and y axis limitations


# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter");
# Customizing ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol");

ax0.set_xlim([50, 80])


ax0.legend(*scatter.legend_elements(), title="Target")

#Add Mean Line
ax0.axhline(over_50["chol"].mean())
#
#

#Add Data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter");
# Customizing ax1
ax1.set(title="Heart Disease and Heart Rate",
        xlabel="Age",
        ylabel="Heart Rate");

ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

ax1.legend(*scatter.legend_elements(), title="Target")

#Add Mean Line
ax1.axhline(over_50["thalach"].mean())

# Add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
